```python
import requests
from bs4 import BeautifulSoup as bs
import pandas as pd
import numpy as np
import pandas as pd
```

```python
url = "https://www.imdb.com/search/title/?groups=top_1000&sort=user_rating,desc"
headers = {"Accept-Language": "en-US, en;q=0.5"}
results = requests.get(url, headers=headers)
urls =[]
```

```python
soup = bs(results.text, "html.parser")
```

```python
movie_div = soup.find_all('div', class_='lister-item-content')
```

```python
for container in movie_div:
    name = container.h3.a['href']
    urls.append(name)
```

```python
for i in range(21):
    url = f"https://www.imdb.com/search/title/?groups=top_1000&sort=user_rating,desc&start={(i+1)*50+1}&ref_=adv_nxt"
    headers = {"Accept-Language": "en-US, en;q=0.5"}
    results = requests.get(url, headers=headers)
    soup = bs(results.text, "html.parser")
    movie_div = soup.find_all('div', class_='lister-item-content')
    for container in movie_div:
        name = container.h3.a['href']
        urls.append(name)
```

```python
# save the 1000 movie urls to a txt file

f=open('top1000-urls.txt','w')
for ele in urls:
    f.write(ele+'\n')
f.close()
```

```python
def find_info(title_url):
    url = f"https://www.imdb.com{title_url}?ref_=adv_li_tt"
    headers = {"Accept-Language": "en-US, en;q=0.5"}
    results = requests.get(url, headers=headers)
    soup = bs(results.text, "html.parser")
    title=""
    image=""
    release_date=""
    country=""
    budget=""
    CWG=""
    production_co=""
    runtime=""
    storyline=""
    metascore=""
    imdbscore =""
    writer =[]
    director =[]
    cast = []
    genres = []
    keywords = []


    poster = soup.find_all('div', class_='poster')
    for container in poster:
        image = container.a.img['src']



    director_writers = soup.find_all('div', class_='credit_summary_item')
    for container in director_writers:
        if container.h4.text == "Director:":
            for elem in container.find_all("a"):
                director.append(elem.text)
        if container.h4.text == "Writers:":
            for elem in container.find_all("a"):
                writer.append(elem.text)



    div = soup.find('div', {"id":"titleCast"})
    i=0
    for elem in div.find_all("a"):
        if i%3:
            name = elem.text
            cast.append(name.rstrip('\n').lstrip(' '))
        i +=1
    cast = list(filter(None, cast[1:-2]))



    div = soup.find_all('div', class_='txt-block')
    for container in div:
        try:
            if container.h4.text == "Release Date:":
                release_date = container.text[container.text.index(':')+2:].split('\n')[0]
            elif container.h4.text == "Country:":
                country = container.a.text
            elif container.h4.text == "Budget:":
                budget = container.text[container.text.index(':')+1:].split('\n')[0].replace('\n','')
            elif container.h4.text == "Cumulative Worldwide Gross:":
                CWG = container.text[container.text.index(':')+2:]
            elif container.h4.text == "Production Co:":
                production_co = container.a.text[1:]
            elif container.h4.text == "Runtime:":
                runtime = container.text[container.text.index(':')+2:].split('\n')[0]

        except:
            pass


    div = soup.find_all('div', class_='inline canwrap')
    for container in div:
        storyline = container.p.text.replace('\n',' ')



    div = soup.find_all('div', class_='see-more inline canwrap')
    for container in div:
        if container.h4.text == "Genres:":
            for elem in container.find_all("a"):
                genres.append(elem.text[1:])
        if container.h4.text == "Plot Keywords:":
            for elem in container.find_all("a"):
                keywords.append(elem.text[1:])
    keywords = keywords[:-1]


    div = soup.find_all('div', class_='metacriticScore score_favorable titleReviewBarSubItem')
    for container in div:
        metascore = container.span.text

    div = soup.find_all('div', class_='ratingValue')
    for container in div:
        imdbscore = container.span.text

    div = soup.find_all('div', class_='title_wrapper')
    for container in div:
        title = container.h1.text.split('\xa0')[0]

    return {"title" : title, "image":image, "director":director, "writer":writer, "cast":cast, "release_date":release_date, "country":country, "budget":budget, "CWG":CWG, "production_co":production_co, "runtime":runtime, "storyline":storyline, "genres":genres, "keywords":keywords, "metascore":metascore, "imdbscore":imdbscore}
```

```python
columns_names = ["id","title", "image", "director", "writer", "cast", "release_date", "country", "budget", "CWG", "production_co", "runtime", "storyline", "genres", "keywords", "metascore", "imdbscore","imdb_url"]
movies = pd.DataFrame(columns = columns_names)
movies.head()
```

```python
for i in range(len(urls)):
    title_url = urls[i]
    movie = find_info(title_url)
    movie['id']=i
    movie['imdb_url']= title_url
    movies = movies.append(movie, ignore_index = True)
```
